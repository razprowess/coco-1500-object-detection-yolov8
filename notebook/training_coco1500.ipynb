```python
import shutil
import os

# Move everything up one level to remove nesting
old_path = "/content/coco_1500_yolo/coco_1500_yolo"
new_path = "/content/coco_1500_yolo_fixed"

if os.path.exists(old_path):
    shutil.move(old_path, new_path)
    print("✅ Fixed folder structure!")
    print(os.listdir(new_path))
else:
    print("Path not found")
```

```python
import yaml

yaml_path = "/content/coco_1500_yolo_fixed/data.yaml"

with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Update paths to match Colab structure
data['train'] = "/content/coco_1500_yolo_fixed/images/train"
data['val']   = "/content/coco_1500_yolo_fixed/images/train"

# Save the updated file
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("✅ data.yaml fixed!")
print("Updated content:")
with open(yaml_path, 'r') as f:
    print(f.read())
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/coco_1500_yolo_fixed/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16,
    name="coco_1500_yolov8",
    patience=10,
    save=True,
    exist_ok=True,
    device=0
)
```
